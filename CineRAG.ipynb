**The Subtitles are zipped. So we need to unzip it first**

```python
import zipfile
import io

# Unzipping the Subs
def extract_subtitle(binary_data):
    zip_file = zipfile.ZipFile(io.BytesIO(binary_data))
    subtitle = zip_file.read(zip_file.namelist()[0]).decode("latin-1")
    return subtitle
```

### Cleaning the subtitles

```python
import re

def clean_subtitle(text):

    text = str(text)

    # Remove timestamps
    text = re.sub(r'\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}', ' ', text)

    # Remove subtitle index numbers
    text = re.sub(r'\n?\d+\n', ' ', text)

    # Remove HTML tags like <i>, </i>
    text = re.sub(r'<.*?>', ' ', text)

    # Remove "Synced by ..." lines
    text = re.sub(r'synced\s*by\s*\w+', ' ', text, flags=re.I)

    # Remove OpenSubtitles ads
    text = re.sub(r'watch any video online with open[- ]?subtitles', ' ', text, flags=re.I)
    text = re.sub(r'free browser extension', ' ', text, flags=re.I)
    text = re.sub(r'please rate this subtitle.*', ' ', text, flags=re.I)
    text = re.sub(r'advertise your product or brand here.*', ' ', text, flags=re.I)
    text = re.sub(r'contact\s*www\.opensubtitles\.org.*', ' ', text, flags=re.I)
    text = re.sub(r'opensubtitles', ' ', text, flags=re.I)
    text = re.sub(r'osdb\.link\S*', ' ', text)

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', ' ', text)

    # Remove weird encoding characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)

    # Remove special characters
    text = re.sub(r"[^a-zA-Z\s']", ' ', text)

    # Fix contractions
    text = re.sub(r"\s+'\s+", "'", text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    return text.lower().strip()
```

### Cleaning Name 

```python
def clean_name(name):

    name = name.lower()

    # extract year
    year_match = re.search(r'\((\d{4})\)', name)
    year = year_match.group(1) if year_match else ""

    # extract season
    season_match = re.search(r's\d+', name)
    season = season_match.group(0) if season_match else ""

    # extract episode
    episode_match = re.search(r'e\d+', name)
    episode = episode_match.group(0) if episode_match else ""

    # remove unwanted parts
    name = re.sub(r'\(\d{4}\)', '', name)      # remove (year)
    name = re.sub(r'eng', '', name)            # remove eng
    name = re.sub(r'\d+cd', '', name)          # remove 1cd
    name = re.sub(r's\d+.*', '', name)         # remove everything after season
    name = name.replace('.', ' ')              # replace dots

    # remove extra spaces
    name = re.sub(r'\s+', ' ', name).strip()

    # rebuild clean name
    clean = f"{name} {season} {episode} {year}"
    clean = re.sub(r'\s+', ' ', clean).strip()

    return clean
```

```python
from langchain_text_splitters  import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
all_chunks = []

for idx, row in data.iterrows():

    subtitle_id = row["num"]        # subtitle id
    name = row["clean_name"]              # movie / show name
    text = row["clean_text"]        # cleaned subtitle

    # skip very small subtitles
    if len(text) < 200:
        continue

    chunks = text_splitter.split_text(text)

    for chunk in chunks:
        all_chunks.append({
            "subtitle_id": subtitle_id,
            "name": name,
            "text": chunk
        })

    if idx % 1000 == 0:
        print("Processed subtitles:", idx)
```

### Inserting Vectors into ChromaDB

```python
import chromadb
import numpy as np
import pandas as pd
from tqdm import tqdm

chunked_data = pd.read_csv("subtitle_chunks.csv")

client = chromadb.PersistentClient(path="movie_chroma_db")

collection = client.get_or_create_collection(name="movie_subtitles")

batch_size = 300000 # per batch size
insert_batch = 1000  #Limit to insert vectors in one go is 5200

for i in tqdm(range(0, len(chunked_data), batch_size)):

    emb_file = f"Embeddings/emb_batch_{i}.npy"
    embeddings = np.load(emb_file)

    batch_df = chunked_data.iloc[i:i+batch_size]

    texts = batch_df["text"].tolist()

    metadata = [
        {"subtitle_id": int(row.subtitle_id), "name": row.name}
        for row in batch_df.itertuples()
    ]

    ids = [str(j) for j in range(i, i + len(batch_df))]

    for j in range(0, len(texts), insert_batch):

        collection.add(
            embeddings=embeddings[j:j+insert_batch].tolist(),
            documents=texts[j:j+insert_batch],
            metadatas=metadata[j:j+insert_batch],
            ids=ids[j:j+insert_batch]
        )

    print(f"Inserted embedding batch {i}")
```

### Input for the System

```python
import numpy as np
from sentence_transformers import SentenceTransformer

# Load Model
print("Loading embedding model...")
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded successfully.")


# Preprocess Query
def preprocess(text: str):
    return text.strip().lower()


# Rerank Documents
def rerank(query_embedding, documents, embeddings=None, final_k=5, threshold=0.40):

    if embeddings is None:
        print("Encoding retrieved documents...")
        doc_embeddings = model.encode(documents, normalize_embeddings=True)
    else:
        print("Using embeddings returned by ChromaDB")
        doc_embeddings = np.array(embeddings)

    # cosine similarity
    scores = np.dot(doc_embeddings, query_embedding)

    # sort scores
    sorted_idx = np.argsort(scores)[::-1]

    # filter by threshold
    filtered_idx = [i for i in sorted_idx if scores[i] >= threshold]

    # take top k
    ranked_idx = filtered_idx[:final_k]

    return ranked_idx, scores


# Display Results
def display_results(docs, metas, ranked_idx, scores):

    if not ranked_idx:
        print("\nNo results passed the similarity threshold.")
        return

    print("\nTop Matching Dialogues\n")

    for rank, i in enumerate(ranked_idx, 1):

        movie = metas[i].get("name", "Unknown")
        subtitle_id = metas[i].get("subtitle_id", "N/A")
        score = round(float(scores[i]), 4)

        print(f"Rank        : {rank}")
        print(f"Movie       : {movie}")
        print(f"Subtitle ID : {subtitle_id}")
        print(f"Score       : {score}")
        print(f"Dialogue    : {docs[i][:300]}")
        print("-" * 60)


# Search Function
def search_movie_dialogue(collection, retrieve_k=100, final_k=5, threshold=0.55):

    print("\n--- Movie Dialogue Search ---")

    query = input("Enter movie dialogue or scene: ")

    if not query.strip():
        print("Query cannot be empty.")
        return

    print("\nProcessing query...")

    query = preprocess(query)
    query_embedding = model.encode(query, normalize_embeddings=True)

    print(f"Retrieving top {retrieve_k} documents from database...")

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=retrieve_k,
        include=["documents", "metadatas", "embeddings"]
    )

    docs = results.get("documents", [[]])[0]
    metas = results.get("metadatas", [[]])[0]
    embeddings = results.get("embeddings", [[]])[0]

    if not docs:
        print("No results found.")
        return

    ranked_idx, scores = rerank(
        query_embedding,
        docs,
        embeddings=embeddings,
        final_k=final_k,
        threshold=threshold
    )

    display_results(docs, metas, ranked_idx, scores)

    return query, docs, metas, ranked_idx
```

#### Defining Prompt

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

prompt = [
    ("system",
     "You are a movie dialogue assistant. "
     "Use ONLY the provided context to answer the question. "
     "Explain properly what is happening the given scene"
     "If the context does not contain the answer, say you do not know."
     "If the user query does not match the context, say do you mean 'the context' and explain the context"),

    MessagesPlaceholder(variable_name="chat_history"),

    ("human",
     "Context:\n{context}\n\n"
     "Question:\n{user_input}\n\n"
     "Answer the question clearly and mention the movie if possible.")
]

prompt_temp = ChatPromptTemplate.from_messages(prompt)
```

#### Output Parser

```python
from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
```

#### History

```python
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
memory={}
def get_session_history(session_id: str):
    if session_id not in memory:
        memory[session_id] = InMemoryChatMessageHistory()
    return memory[session_id]
```

#### Chain

```python
base_chain = prompt_temp | chat_model | output_parser

chain = RunnableWithMessageHistory(
    base_chain,
    get_session_history,
    input_messages_key="user_input",
    history_messages_key="chat_history",
)
```

### Pipeline

```python
def rag_pipeline(collection):

    context = []

    print('Searching movie through our Database....................')

    result = search_movie_dialogue(collection)

    if result is None:
        return

    query, docs, metas, ranked_idx = result

    print('Building Context for LLM....................')

    for rank, i in enumerate(ranked_idx, 1):

        movie = metas[i].get("name", "Unknown")
        dialogue = docs[i]

        context.append(
            f"""Document {rank}
            Movie: {movie}
            Dialogue: {dialogue}
            """
                    )

    context_text = "\n".join(context)

    response = chain.invoke(
        {
            "user_input": query,
            "context": context_text
        },
        config={"configurable": {"session_id": "user1"}}
    )

    print("\nAnswer:\n")
    print(response)
```
